```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
os.chdir("../csv file")
```

```python
data = pd.read_csv("globalterrorismdb_0718dist.tar.bz2" , compression="bz2")
data
```

```python
main_data = data[["iyear" ,"region_txt" ,"country_txt" , "success" , "suicide" , "attacktype1" , "attacktype1_txt" , "targtype1_txt" , "targsubtype1_txt" , "target1" ,
      "natlty1_txt" , "gname" , "gsubname" , "nperps" , "weaptype1_txt" , "weapsubtype1_txt" , "nkill" , "nkillus"]]
main_data
```

# 1- How has the Number of Terrorist activities changed over the Years? Are there certain Regions where this trend is different from the global Averages?

```python
main_data["target1"].unique()
```

```python
fig, ax = plt.subplots(figsize=(16, 7))
ax.hist(main_data["iyear"] , bins=40 , color="#ff6347")
ax.set_xlabel("Year" , color = "gray" , loc = "left" , fontsize = 14 , weight = "bold")
ax.set_ylabel("Number of Attacks" , loc = "bottom" , fontsize = 14 , color ="gray" , weight = "bold")
ax.set_title("Number of Attacks in different Years" , loc = "left" , fontsize = 14 , color ="gray" , weight = "bold")
ax.tick_params(axis = "x" , colors = "gray", labelsize="large")
ax.tick_params(axis = "y" , colors = "gray", labelsize="large")
ax.spines[["right", "top"]].set_visible(False)
ax.spines[["left", "bottom"]].set_color("gray")
plt.show()
```

```python
count_of_atacks = main_data.groupby("region_txt" , as_index=False).agg({"success" : "count"}).sort_values("success")
count_of_atacks
```

## In the Middle East & North Africa and South Asia, the number of attacks is much higher than the global average.
## In Australasia & Oceania and Central Asia and East Asia, the number of attacks is much lower than the global average.

```python
fig , ax = plt.subplots(figsize = (16,7))
ax.barh(count_of_atacks["region_txt"] , count_of_atacks["success"] ,
    color= np.where(count_of_atacks["success"].gt(20000), "#ff6347", np.where(count_of_atacks["success"].lt(3000), "green", "gray")))
ax.vlines(count_of_atacks["success"].mean() , ymin=-.5 , ymax=11.5, color= "blue" , linewidth = 4 , linestyles="dashed")
ax.set_xlabel("Number of Atacks" , color = "gray" , loc = "left" , fontsize = 14 , weight = "bold")
ax.set_ylabel("Region" , loc = "bottom" , fontsize = 14 , color ="gray" , weight = "bold")
ax.set_title("Number of Attacks in different Regions" , loc = "left" , fontsize = 14 , color ="gray" , weight = "bold")
ax.tick_params(axis = "x" , colors = "gray", labelsize="large")
ax.tick_params(axis = "y" , colors = "gray", labelsize="large")
ax.spines[["right", "top"]].set_visible(False)
ax.spines[["left", "bottom"]].set_color("gray")
plt.show()
```

# 2- Is the Number of Incidents and the Number of Casualties correlated? Can you Spot any Irregularities or Outliers?

```python
number_of_kills = main_data.groupby("iyear" , as_index=False).agg({"nkill" : "count"})
number_of_kills
```

# 3- What are the most common Methods of Attacks? Does it differ in Various Regions or in Time?

```python
atack_type = main_data.groupby("attacktype1_txt" , as_index=False).agg({"iyear" : "count"}).sort_values("iyear")
atack_type
```

# The most common Method of Attack is Bombing/Explosion, which is more than the other Methods by a large Margin.

```python
fig , ax = plt.subplots(figsize = (16,7))
ax.barh(atack_type["attacktype1_txt"] , atack_type["iyear"] ,
    color =np.where(atack_type["iyear"].gt(40000), "#ff6347", np.where(atack_type["iyear"].lt(5000), "green", "gray")))
ax.set_xlabel("Number of Attacks" , color = "gray" , loc = "left" , fontsize = 14 , weight = "bold")
ax.set_ylabel("Methods of Attacks" , loc = "bottom" , fontsize = 14 , color ="gray" , weight = "bold")
ax.set_title("Number of Attacks in different Methods" , loc = "left" , fontsize = 14 , color ="gray" , weight = "bold")
ax.vlines(atack_type["iyear"].mean() , ymin=-.5 , ymax=8.5 , color = "blue" , linewidth = 4 , linestyles="dashed")
ax.tick_params(axis = "x" , colors = "gray", labelsize="large")
ax.tick_params(axis = "y" , colors = "gray", labelsize="large")
ax.spines[["right", "top"]].set_visible(False)
ax.spines[["left", "bottom"]].set_color("gray")
plt.show()
```

# 5 - Which Groups have carried out the most Attacks?

```python
maximum_attack = main_data.groupby("gname" , as_index=False).agg({"iyear" : "count"}).sort_values("iyear").tail(11)
maximum_attack = maximum_attack[maximum_attack["gname"].ne("Unknown")]
maximum_attack
```

```python
fig , ax = plt.subplots(figsize = (16,7))
ax.barh(maximum_attack["gname"] , maximum_attack["iyear"] ,
    color =np.where(maximum_attack["iyear"].gt(5500), "#ff6347", np.where(maximum_attack["iyear"].lt(3000), "green", "gray")))
ax.set_xlabel("Number of Attacks" , color = "gray" , loc = "left" , fontsize = 14 , weight = "bold")
ax.set_ylabel("Group Name" , loc = "bottom" , fontsize = 14 , color ="gray" , weight = "bold")
ax.set_title("Maximum Atacks of Different Groups" , loc = "left" , fontsize = 14 , color ="gray" , weight = "bold")
ax.tick_params(axis = "x" , colors = "gray", labelsize="large")
ax.tick_params(axis = "y" , colors = "gray", labelsize="large")
ax.spines[["right", "top"]].set_visible(False)
ax.spines[["left", "bottom"]].set_color("gray")
plt.show()
```
